```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Ensure hr_data is a Pandas DataFrame
hr_data = pd.DataFrame(hr_data)

# Select only numeric columns
hr_data_numeric = hr_data.select_dtypes(include=[np.number])

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(20, 20))

# Create the heatmap
sns.heatmap(hr_data_numeric.corr(), annot=True, linewidths=0.5, fmt='.1f', ax=ax)

# Show the plot
plt.show()
```

```python
#plot distributions
k=1
plt.figure(figsize=(40, 40))
for col in hr_data:
  if col=="Attrition":
    continue
  yes = hr_data[hr_data['Attrition'] == 'Yes'][col]
  no = hr_data[hr_data['Attrition'] == 'No'][col]
  plt.subplot(6, 6, k)
  plt.hist(yes, bins=25, alpha=0.5, label='yes', color='b')
  plt.hist(no, bins=25, alpha=0.5, label='no', color='r')
  plt.legend(loc='upper right')
  plt.title(col)
  k+=1
```
